```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(data=tips,x='sex')
```

```python
sns.set_style('white')
sns.countplot(data=tips,x='sex')
```

```python
sns.set_style('ticks')
sns.countplot(data=tips,x='sex',palette='deep')
```

```python
#SPine removal
sns.countplot(data=tips,x='sex')
sns.despine()
```

```python
sns.countplot(data=tips,x='sex')
sns.despine(left=True)
```

```python
#Size and aspect
```

```python
plt.figure(figsize=(12,3))
sns.countplot(data=tips,x='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',size=2,aspect=4)
```

```python
#Scale adn COntext
```

```python
sns.set_context('poster',font_scale=4)
sns.countplot(data=tips,x='sex')
```
