```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',palette='coolwarm')
```

```python
#markers
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v']
          ,scatter_kws={'s':100})
```

```python
#Using a grid
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='sex',col='time')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',hue='sex')
```

```python
#ASpectandSize
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',hue='sex',aspect=0.7,size=8)
```
