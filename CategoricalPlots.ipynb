```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(data=tips,y='total_bill',x='sex')
```

```python
import numpy as np
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',estimator=np.std)
```

```python
sns.countplot(data=tips,x='sex')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',palette='rainbow')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.boxplot(data=tips,orient='h')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='smoker')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill')
```
